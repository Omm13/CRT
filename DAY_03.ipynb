````python
# Factorial
def factorial(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  print(fact)
factorial(5)
````

````python
# Recursion Backwards
def show(n):
  if n == 0:
    return
  print(n)
  show(n-1)
show(5)
````

````python
# Recursion Forwards
def show(n):
  for i in range(1, n+1):
    if i == n+1:
      return
    print(i)
show(10)
````

````python
# Recursion Learning
def learningstack(n):
  if(0==n):
    return
  print(f'inside fn, before: {n}')
  learningstack(n-1)
  print(f'inside fn, after: {n}')

n = 3
print(f'Code Starts {n}')
learningstack(n)
print(f'Learnt Call Stacking working {n}')
````

````python
# Sum of natural numbers by Using Loops
def sum(n):
  sum = 0
  for i in range(1, n+1):
    sum = sum + i
  print(sum)
sum(10)
````

````python
# Sum of natural Numbers by Using recursion
def summation(n,sum=0):
  if n == 0:
    print(sum)
    return
  summation(n-1,sum + n)
summation(5)

# Factorial using Recursion
def fact(m , f=1):
  if m == 0:
    print(f)
    return
  fact(m-1, f*m)
fact(5)
````

````python
# Create student Class, takes name & Marks, Creat a method of average and print Average
class Student:
  def __init__(self, name, m1, m2, m3):
    self.name = name
    self.m1 = m1
    self.m2 = m2
    self.m3 = m3

  def printAvg(self):
    avg = (self.m1 + self.m2 + self.m3)/3
    print(f'Avg of {self.name} = {avg:.2f}')

s1 = Student("Omm", 85, 80, 90)
s1.printAvg()
````

```python
# Factorial
def factorial(n):
  fact = 1
  for i in range(1, n+1):
    fact = fact * i
  print(fact)
factorial(5)
```

```python
# Recursion Backwards
def show(n):
  if n == 0:
    return
  print(n)
  show(n-1)
show(5)
```

```python
# Recursion Forwards
def show(n):
  for i in range(1, n+1):
    if i == n+1:
      return
    print(i)
show(10)
```

```python
# Recursion Learning
def learningstack(n):
  if(0==n):
    return
  print(f'inside fn, before: {n}')
  learningstack(n-1)
  print(f'inside fn, after: {n}')

n = 3
print(f'Code Starts {n}')
learningstack(n)
print(f'Learnt Call Stacking working {n}')
```

```python
# Sum of natural numbers by Using Loops
def sum(n):
  sum = 0
  for i in range(1, n+1):
    sum = sum + i
  print(sum)
sum(10)
```

```python
# Sum of natural Numbers by Using recursion
def summation(n):
  sum = 0
  if n == n:
    sum = sum + n
    return
  print(sum)
summation(10)
```

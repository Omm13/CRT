```python
# Sum of Numbers
num = 1234
sum = 0
while num!=0:
  rem = num % 10
  sum = sum + rem
  num = num // 10
print(sum)
```

```python
# Reverse a number
num = 1234
rev = 0
while num!=0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print(rev)
```

```python
# Reverse a number
num = 121
n = num
rev = 0
while num!=0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10
print(rev)
if n == rev:
  print("It is Palindrome")
else:
  print("It is not palindrome")
```

```python
# Armstrong Number
num = int(input("enter a number: "))
t=num
l=len(str(num))
temp = 0
while num!=0:
  rem = num % 10
  temp = temp+rem**l
  num //= 10
print(temp)
if temp == t:
  print("armstrong")
else:
  print("not Armstrong")
```

```python
arr = [11, 88, 22, 34, 67, 89, 20]
even = []
odd = []
for i in range(len(arr)):
  if arr[i]%2==0:
    even.append(arr[i])
  else:
    odd.append(arr[i])
print(f'odd : {odd}')
print(f'even: {even}')
```

```python
# Replace 10 with 100
# Using Iterative method
arr = [10, 20, 30, 10, 50, 60, 10, 90, 10]
key = 100
for i in range(len(arr)):
  if arr[i]==10:
    arr[i]=key
print(arr)

# Only 1st Occurences changing
arr = [10, 20, 30, 10, 50, 60, 10, 90, 10]
key = 100
for i in range(len(arr)):
  if arr[i]==10:
    arr[i]=key
    break
print(arr)

# Last occurences
arr = [10, 20, 30, 10, 50, 60, 10, 90]
key = 100
for i in range(len(arr)-1,-1,-1):
  if arr[i]==10:
    arr[i]=key
    break
print(arr)
```

```python
# Reverse Using Two Pointers
array = [10, 20, 30, 10, 50, 60, 10, 90, 10]
start = 0
end = len(array)-1
while (start < end):
  array[start], array[end] = array[end], array[start]
  start+=1
  end-=1
print(array)
```
